```python
# Constants
COST_C = 3
COST_S = 2
COST_G = 2
SPECIAL_PACK_DISCOUNT_PRICE = 10
```

```python
# Function to check if the weight is valid for a given product
def is_valid_weight(product, weight):
    if product == 'c':
        return 24.9 <= weight <= 25.1
    elif product == 's':
        return 49.9 <= weight <= 50.1
    elif product == 'g':
        return 49.9 <= weight <= 50.1
    else:
        return False
```

```python
# Define product costs
product_costs = {'C': COST_C, 'S': COST_S, 'G': COST_G}
```

```python
# Function to process customer order
def process_customer_order():
    try:
        # Receive customer order
        order = {
            'C': int(input("Enter the number of bags of C: ")),
            'S': int(input("Enter the number of bags of S: ")),
            'G': int(input("Enter the number of bags of G: "))
        }

        # Calculate the number of special pack discounts
        num_special_pack_discounts = min(order['C'], order['S'] // 2, order['G'] // 2)

        # Calculate total cost before discount
        total_cost_before_discount = sum(order[product] * product_costs[product] for product in order)

        # Calculate cost of remaining bags after discount
        cost_of_remaining_bags = (
            (order['C'] - num_special_pack_discounts) * product_costs['C'] +
            (order['S'] - (2 * num_special_pack_discounts)) * product_costs['S'] +
            (order['G'] - (2 * num_special_pack_discounts)) * product_costs['G']
        )

        # Calculate total cost after applying special pack discount
        total_cost_after_discount = (num_special_pack_discounts * SPECIAL_PACK_DISCOUNT_PRICE) + cost_of_remaining_bags

        # Calculate money saved with special pack discount
        money_saved = total_cost_before_discount - total_cost_after_discount

        # Print or use the calculated values
        print("Total Cost Before Discount:", total_cost_before_discount)
        print("Total Cost After Discount:", total_cost_after_discount)
        print("Money Saved:", money_saved)

    except ValueError:
        print("Invalid input. Please enter valid integers for the number of bags.")
```

```python
# Main menu
while True:
    print("\nMain Menu:")
    print("1. Check weights for delivery")
    print("2. Enter customer order")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == '1':
        product = input("Enter the product (C/S/G): ").lower()
        weight = float(input("Enter the weight of the bag: "))
        if is_valid_weight(product, weight):
            print("Weight is valid.")
        else:
            print("Invalid weight. Please check the weight requirements for the product.")
    elif choice == '2':
        process_customer_order()
    elif choice == '3':
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 3.")
```
